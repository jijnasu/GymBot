## Objective : 
####Given a dataset about different conditions of a patient and taking it into account we have to train a model that will predict whether the chances of getting a heart attack is high or low. 

# Part A : Data preprocessing
##### Step 1 : Importing libraries


```python
import numpy as np  #for numeric calculation
import pandas as pd #for working with data
import tensorflow as tf #deep learning framework
```

```python
dataset = pd.read_csv("Heart.csv") #reading the dataset
```

```python
dataset #visualize the dataset
```

##### Step 2: To create Feature Matrix(x) and Dependent variable vector(y)

```python
x=dataset.iloc[:,:-1].values 
y=dataset.iloc[:,-1:].values #it is dependent on x
```

```python
y
```

##### Step 3: Replacing Missing Data(by using the average)

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean') #the missing data is nan(not a number)
imputer.fit(x)
x=imputer.transform(x)
```

##### Step 4: Splitting the dataset into training and test sets

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
 #we r dividing data into train n test set randomly in ratio 8:2
```

##### Step 5: Scaling the dataset (to avoid baised model towards larger value)

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.fit_transform(xtest)
```

```python
xtest
```

# Part B: Build the Model
##### Step 1: Training the model

```python
t1=tf.keras.models.Sequential()
```

```python
#adding a hidden layer
# t1.add(tf.keras.layers.Dense(units=4,activation='relu'))
#adding a hidden layer
# t1.add(tf.keras.layers.Dense(units=2,activation='relu'))
# #adding a hidden layer
# t1.add(tf.keras.layers.Dense(units=2,activation='relu'))
#o/p layer
t1.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

##### Step 2: compile the model

```python
t1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

##### Step 3: training the model

```python
t1.fit(xtrain,ytrain,epochs=200)
```

##### Step 4: testing model

```python
y1=t1.predict(xtest)
```

```python
y1
```

##### Step 5: Converting the output into boolean (true if >0.5 and false if less)

```python
y1=(y1>0.5)
y1
```

```python
from sklearn.preprocessing import LabelEncoder #Encoding the true false as 1 and 0
le= LabelEncoder()
y1=le.fit_transform(y1)
y1
```

```python
print(np.concatenate((y1.reshape(len(y1),1),ytest.reshape(len(ytest),1)),1))
# to visualize the ytest and y predicted side-by-side
```

##### Step 6:  performance matrix ( to know the accuracy )

```python
from sklearn.metrics import confusion_matrix, accuracy_score , precision_score 
cm= confusion_matrix(ytest,y1)
print(cm)
print(accuracy_score(ytest,y1))
print(precision_score(ytest,y1))
```
